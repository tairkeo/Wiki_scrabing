```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find_all('table')[1]
titles = table.find_all('th')
```

```python
wtitles = [title.text.strip() for title in titles]
print(wtitles)
    
    
```

```python
df = pd.DataFrame(columns = wtitles)
```

```python
raw_lines = table.find_all('tr')
for raws in raw_lines[1:]:
    raw_lines2 = raws.find_all('td')
    Info = [raws2.text.strip() for raws2 in raw_lines2 ]

    lenght = len(df)
    df.loc[lenght] = Info
    
```

```python
pd.set_option('display.max_rows', None)
df = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(df['Revenue growth'])
plt.show()
```
